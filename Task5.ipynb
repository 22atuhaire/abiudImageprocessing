# Image Resizing and Scaling

```python
import cv2
import numpy as np
```

```python
def main():
    try:
        # Loading the Image
        image = cv2.imread("images/image2.jpg")

        if image is None:
            raise FileNotFoundError("Image file not found. Please check the file path.")

        original_height, original_width = image.shape[:2]

        # 1. Resize to Exact Dimensions (400x300)
        resized_exact = cv2.resize(image, (400, 300))

        # 2. Resize to 50% of Original Size (Maintain Aspect Ratio)
        scaled_50_percent = cv2.resize(
            image,
            None,
            fx=0.5,
            fy=0.5,
            interpolation=cv2.INTER_LINEAR
        )

        # Saving the 50% scaled output
        save_status = cv2.imwrite("task5_resized.jpg", scaled_50_percent)

        if not save_status:
            raise Exception("Failed to save 50% resized image.")

        # 3. Resize to Width = 800 Pixels 
        new_width = 800
        aspect_ratio = original_height / original_width
        new_height = int(new_width * aspect_ratio)

        resized_width_800 = cv2.resize(
            image,
            (new_width, new_height)
        )

        # 4. Compare Different Interpolation Methods
        # Resize same image to 400x300 using different methods

        resized_nearest = cv2.resize(
            image,
            (400, 300),
            interpolation=cv2.INTER_NEAREST
        )

        resized_linear = cv2.resize(
            image,
            (400, 300),
            interpolation=cv2.INTER_LINEAR
        )

        resized_cubic = cv2.resize(
            image,
            (400, 300),
            interpolation=cv2.INTER_CUBIC
        )


        # Display All Results
        cv2.imshow("Original Image", image)
        cv2.imshow("Exact 400x300 (Distorted)", resized_exact)
        cv2.imshow("50% Scaled", scaled_50_percent)
        cv2.imshow("Width 800 (Aspect Ratio Maintained)", resized_width_800)
        cv2.imshow("Interpolation - Nearest", resized_nearest)
        cv2.imshow("Interpolation - Linear", resized_linear)
        cv2.imshow("Interpolation - Cubic", resized_cubic)

        cv2.waitKey(0)
        cv2.destroyAllWindows()

        print("Task 5 completed successfully.")

    except FileNotFoundError as file_error:
        print("Error:", file_error)

    except Exception as general_error:
        print("Unexpected error occurred:", general_error)


if __name__ == "__main__":
    main()
```
