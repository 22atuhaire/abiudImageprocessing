# Image Cropping Techniques

```python
import cv2
import numpy as np
import os
```

```python
def main():
    try:
        # Load Image
        original_image = cv2.imread("images/image3.jpg")

        if original_image is None:
            raise FileNotFoundError("Image file not found. Check path.")

        image_height, image_width = original_image.shape[:2]

        # Create copy for drawing rectangles
        display_image = original_image.copy()

        # 1. CENTER CROP (Central 50%)
        center_crop_width = image_width // 2
        center_crop_height = image_height // 2

        start_x = (image_width - center_crop_width) // 2
        start_y = (image_height - center_crop_height) // 2

        end_x = start_x + center_crop_width
        end_y = start_y + center_crop_height

        center_crop = original_image[start_y:end_y, start_x:end_x]

        # Draw rectangle
        cv2.rectangle(display_image, (start_x, start_y), (end_x, end_y), (0, 255, 0), 2)

        cv2.imwrite("task7_center_crop.jpg", center_crop)

        # 2. Square crop (Largest Square from Center)
        square_size = min(image_width, image_height)

        square_start_x = (image_width - square_size) // 2
        square_start_y = (image_height - square_size) // 2

        square_end_x = square_start_x + square_size
        square_end_y = square_start_y + square_size

        square_crop = original_image[square_start_y:square_end_y,
                                     square_start_x:square_end_x]

        cv2.rectangle(display_image,
                      (square_start_x, square_start_y),
                      (square_end_x, square_end_y),
                      (255, 0, 0),
                      2)

        cv2.imwrite("task7_square_crop.jpg", square_crop)

        # -------------------------------------------------
        # 3. ROI Crop (100,100) to (400,400)
        # -------------------------------------------------
        roi_start_x, roi_start_y = 100, 100
        roi_end_x, roi_end_y = 400, 400

        # Ensure ROI does not exceed image boundaries
        roi_end_x = min(roi_end_x, image_width)
        roi_end_y = min(roi_end_y, image_height)

        roi_crop = original_image[roi_start_y:roi_end_y,
                                  roi_start_x:roi_end_x]

        cv2.rectangle(display_image,
                      (roi_start_x, roi_start_y),
                      (roi_end_x, roi_end_y),
                      (0, 0, 255),
                      2)

        cv2.imwrite("task7_roi_crop.jpg", roi_crop)

        
        # 4. GRID CROP (3x3)
    
        cell_height = image_height // 3
        cell_width = image_width // 3

        for row in range(3):
            for col in range(3):
                y_start = row * cell_height
                y_end = (row + 1) * cell_height
                x_start = col * cell_width
                x_end = (col + 1) * cell_width

                grid_crop = original_image[y_start:y_end, x_start:x_end]

                filename = f"task7_grid_row{row+1}_col{col+1}.jpg"
                cv2.imwrite(filename, grid_crop)

        
        # Display Original with Rectangles
        cv2.imshow("Original with Crop Areas", display_image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

        print("Task 7 completed successfully.")

    except FileNotFoundError as file_error:
        print("Error:", file_error)

    except Exception as general_error:
        print("Unexpected error occurred:", general_error)


if __name__ == "__main__":
    main()
```
