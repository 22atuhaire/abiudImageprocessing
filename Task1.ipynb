# 1. Image Loading and Analysis

```python
import cv2
import numpy as np
import os
```

```python
def read_image(image_name):
    """
    Read an image from the images directory using OpenCV.
    Args:
        image_name (str): Name of the image file including extension 
    Returns:
        numpy.ndarray: Image as a numpy array, or None if loading fails
    """
    
    try:
        # Construct the full file path and read the image using OpenCV
        image = cv2.imread(f"images/{image_name}")
        
        #check if the image was read 
        if image is None:
            raise ValueError(f"Could not read the image at {image_name}")
        return image
    
    except FileNotFoundError:

        # Handle case when the image file does not exist
        print(f"Image file {image_name} not found.")
        return None
    except Exception as e:
        
        # Handle any other unexpected errors during image loading
        print(f"An error occurred: {e}")
        return None
```

```python
# Load the image from the images directory
image2 = read_image("image2.jpg")

# Check if the image was successfully loaded
if image2 is not None:

    # Extract the image dimensions from the shape attribute
    #where channels = 3 (BGR)
    height, width, channels = image2.shape
    
    # Print each dimension on a separate line for clarity
    print(f"Height: {height}")
    print(f"Width: {width}")
    print(f"Channels: {channels}")
else:
    # Print error message if image loading failed
    print("Failed to load image")
```

# 2. Data type of the image array

```python
# Display the data type of the image array
if image2 is not None:
    print(f"Data type: {image2.dtype}")
else:
    print("No image loaded")
```

# 3. Total number of pixels

```python
# Calculate the total number of pixels in the image
if image2 is not None:
    total_pixels = height * width
    print(f"Total number of pixels: {total_pixels:,}")
else:
    print("No image loaded")
```
