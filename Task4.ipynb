# Color Space Conversion

```python
import cv2
import numpy as np
```

```python
def main():
    try:
       
        # Load the same image used in Task 3
        
        image = cv2.imread("images/image2.jpg")

        if image is None:
            raise FileNotFoundError("Image file not found. Please check the path.")

        # 1. Convert BGR to Grayscale
        grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # 2. Convert BGR to HSV
        hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

        # 3. Convert BGR to LAB
        lab_image = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

        # Convert grayscale to 3-channel for stacking
        grayscale_bgr = cv2.cvtColor(grayscale_image, cv2.COLOR_GRAY2BGR)

        # Resize all images to same size (safety measure)
        image_height, image_width = image.shape[:2]

        hsv_resized = cv2.resize(hsv_image, (image_width, image_height))
        lab_resized = cv2.resize(lab_image, (image_width, image_height))
        grayscale_resized = cv2.resize(grayscale_bgr, (image_width, image_height))

        # Create 2x2 Grid
        # Top row: Original  Grayscale
        # Bottom row: HSV | LAB
        top_row = np.hstack((image, grayscale_resized))
        bottom_row = np.hstack((hsv_resized, lab_resized))
        grid_image = np.vstack((top_row, bottom_row))

        # Save Final Grid Image
        save_status = cv2.imwrite("task4_color_spaces.jpg", grid_image)

        if not save_status:
            raise Exception("Failed to save the output image.")

       
        # Display Result
        cv2.imshow("Color Space Conversions (2x2 Grid)", grid_image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

        print("Task 4 completed successfully.")

    
        # (Explanation in Comments)

        """
        BONUS ANSWERS:

        1. Best color space for detecting a red object under varying lighting:
           HSV is best because:
           - Hue represents color independently of brightness.
           - Lighting changes mostly affect the Value channel, not Hue.
           - Therefore, color detection becomes more robust.

        2. Best color space for detecting skin tones:
           LAB is often preferred because:
           - Lightness (L) is separated from color information.
           - The A and B channels represent color components.
           - This separation makes skin detection more stable across lighting variations.
        """

    except FileNotFoundError as file_error:
        print("Error:", file_error)

    except Exception as general_error:
        print("Unexpected error occurred:", general_error)


if __name__ == "__main__":
    main()
```
