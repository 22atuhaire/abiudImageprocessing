# bluring the image

```python
import cv2
import numpy as np
```

```python
def add_salt_and_pepper_noise(image, noise_ratio=0.02):
    """
    Adds salt-and-pepper noise to the image.
    noise_ratio determines the percentage of noisy pixels.
    """
    noisy_image = image.copy()
    height, width = image.shape[:2]

    total_pixels = height * width
    num_noisy_pixels = int(total_pixels * noise_ratio)

    # Add salt (white pixels)
    for _ in range(num_noisy_pixels // 2):
        y = np.random.randint(0, height)
        x = np.random.randint(0, width)
        noisy_image[y, x] = [255, 255, 255]

    # Add pepper (black pixels)
    for _ in range(num_noisy_pixels // 2):
        y = np.random.randint(0, height)
        x = np.random.randint(0, width)
        noisy_image[y, x] = [0, 0, 0]

    return noisy_image


def add_title(image, text):
    """
    Adds title text to image.
    """
    labeled_image = image.copy()
    cv2.putText(
        labeled_image,
        text,
        (20, 40),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (0, 255, 0),
        2,
        cv2.LINE_AA
    )
    return labeled_image
```

```python
def main():
    try:
        # Load Image
        original_image = cv2.imread("images/image3.jpg")

        if original_image is None:
            raise FileNotFoundError("Image file not found. Check path.")

        # 1. Add Salt-and-Pepper Noise
        noisy_image = add_salt_and_pepper_noise(original_image, noise_ratio=0.03)

        # 2. Apply Blurring Methods
        # Average Blur
        average_blur = cv2.blur(noisy_image, (5, 5))

        # Gaussian Blur
        gaussian_blur = cv2.GaussianBlur(noisy_image, (5, 5), 0)

        # Median Blur
        median_blur = cv2.medianBlur(noisy_image, 5)

        # Bilateral Filter
        bilateral_blur = cv2.bilateralFilter(noisy_image, 9, 75, 75)

    
        # 3. Create 2x2 Grid
        # Resize for consistent grid display
        target_size = (400, 300)

        noisy_resized = cv2.resize(noisy_image, target_size)
        gaussian_resized = cv2.resize(gaussian_blur, target_size)
        median_resized = cv2.resize(median_blur, target_size)
        bilateral_resized = cv2.resize(bilateral_blur, target_size)

        # Add titles
        noisy_labeled = add_title(noisy_resized, "Noisy Image")
        gaussian_labeled = add_title(gaussian_resized, "Gaussian Blur")
        median_labeled = add_title(median_resized, "Median Blur")
        bilateral_labeled = add_title(bilateral_resized, "Bilateral Filter")

        # Arrange in 2x2 grid
        top_row = np.hstack((noisy_labeled, gaussian_labeled))
        bottom_row = np.hstack((median_labeled, bilateral_labeled))
        grid_image = np.vstack((top_row, bottom_row))

    
        # Save Output
        save_status = cv2.imwrite("task8_blur_comparison.jpg", grid_image)

        if not save_status:
            raise Exception("Failed to save output image.")

        # Display Result
        cv2.imshow("Task 8 - Blur Comparison", grid_image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

        print("Task 8 completed successfully.")
        # ANALYSIS QUESTIONS (Answer in Comments)

        """
        ANALYSIS ANSWERS:

        1. Which blur method best removes salt-and-pepper noise?
           Median blur performs best because it replaces each pixel
           with the median of its neighborhood, effectively removing
           isolated white and black noise pixels.

        2. Which method best preserves edges while blurring?
           Bilateral filter preserves edges best because it considers
           both spatial proximity and pixel intensity differences,
           preventing smoothing across strong edges.

        3. How does kernel size affect the result?
           - Larger kernel size increases smoothing.
           - More noise is removed.
        """

    except FileNotFoundError as file_error:
        print("Error:", file_error)

    except Exception as general_error:
        print("Unexpected error occurred:", general_error)


if __name__ == "__main__":
    main()
```
