# color channel manipulation

```python
import cv2
import numpy as np
```

```python
def main():
    try:
        #loading the image
        image = cv2.imread('images/image2.jpg')
        if image is None:
            raise FileNotFoundError("The image file was not found.")
        
        #saving the image
        cv2.imwrite('images/task3_original.jpg', image) 

        #split into BGR channels
        b, g, r = cv2.split(image)

        #disabling channels
        zero_channel = np.zeros_like(b)

        #creating image showing only one channel

        #only red channel
        red_image = cv2.merge([zero_channel, zero_channel, r])

        #only green channel
        green_image = cv2.merge([zero_channel, g, zero_channel])

        #only blue channel
        blue_image = cv2.merge([b, zero_channel, zero_channel])

        #Reconstruct Original Image by Combining Channels

        reconstructed_image = cv2.merge([b, g, r])

        #Save All Required Images
        cv2.imwrite('images/task3_red_channel.jpg', red_image)
        cv2.imwrite('images/task3_green_channel.jpg', green_image)
        cv2.imwrite('images/task3_blue_channel.jpg', blue_image)

        #Didplaying the images
        cv2.imshow('Original Image', image)
        cv2.imshow('Red Channel', red_image)
        cv2.imshow('Green Channel', green_image)
        cv2.imshow('Blue Channel', blue_image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

        print("All images have been processed and saved successfully.")

    except FileNotFoundError as file_error:
        print("Error", file_error)

    except Exception as general_error:
        print("An unexpected error occurred:", general_error)   

if __name__ == "__main__":
    main()

    
```
