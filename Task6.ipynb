# Image Rotation

```python
import cv2
import numpy as np
import math
```

```python
def rotate_image_full_view(image, angle):
    """
    Rotating the image without cropping using affine transformation.
    Calculates new bounding dimensions.
    """

    original_height, original_width = image.shape[:2]

    # Compute center of the image
    center = (original_width // 2, original_height // 2)

    # Get rotation matrix
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)

    # Extract cosine and sine values
    cos_value = abs(rotation_matrix[0, 0])
    sin_value = abs(rotation_matrix[0, 1])

    # Compute new bounding dimensions
    new_width = int((original_height * sin_value) + (original_width * cos_value))
    new_height = int((original_height * cos_value) + (original_width * sin_value))

    # Adjust rotation matrix to account for translation
    rotation_matrix[0, 2] += (new_width / 2) - center[0]
    rotation_matrix[1, 2] += (new_height / 2) - center[1]

    # Perform affine transformation
    rotated_image = cv2.warpAffine(image, rotation_matrix, (new_width, new_height))

    return rotated_image

```

```python
def add_label(image, text):
    """
    Adds rotation angle label to image.
    """
    labeled_image = image.copy()
    cv2.putText(
        labeled_image,
        text,
        (20, 40),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (0, 255, 0),
        2,
        cv2.LINE_AA
    )
    return labeled_image


def main():
    try:
        # Load Image
        image = cv2.imread("images/image2.jpg")

        if image is None:
            raise FileNotFoundError("Image file not found. Check path.")

        # Required Rotation Angles
        angles = [30, 60, 90, 135, 180]

        rotated_images = []

        for angle in angles:
            rotated = rotate_image_full_view(image, angle)
            labeled = add_label(rotated, f"{angle} Degrees")
            rotated_images.append(labeled)

        # Resize all rotated images to same size for grid
        target_width = 400
        target_height = 300

        resized_images = []
        for img in rotated_images:
            resized = cv2.resize(img, (target_width, target_height))
            resized_images.append(resized)

        # Add original image as sixth slot
        original_resized = cv2.resize(image, (target_width, target_height))
        original_labeled = add_label(original_resized, "Original")

        # Create 2x3 Grid
        top_row = np.hstack((original_labeled, resized_images[0], resized_images[1]))
        bottom_row = np.hstack((resized_images[2], resized_images[3], resized_images[4]))

        grid_image = np.vstack((top_row, bottom_row))

        # Save Output

        save_status = cv2.imwrite("task6_rotations.jpg", grid_image)

        if not save_status:
            raise Exception("Failed to save output image.")

        # Display Output
        cv2.imshow("Task 6 - Rotations (2x3 Grid)", grid_image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

        print("Task 6 completed successfully.")

    except FileNotFoundError as file_error:
        print("Error:", file_error)

    except Exception as general_error:
        print("Unexpected error occurred:", general_error)


if __name__ == "__main__":
    main()
```
