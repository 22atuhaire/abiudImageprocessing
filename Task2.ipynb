# Creating a blank black image of size 600x600

```python
import cv2
import numpy as np
```

```python
def main():
    try:
        # Create a blank image(3 channels for RGB)
        image_height = 600
        image_width = 600
        image = np.zeros((image_height, image_width, 3), dtype=np.uint8)
    
        # Defining the center points.
        center_cordinates = (image_width // 2, image_height // 2)

        #Outer Circle (Red, radius 250, thickness 5)
        cv2.circle(
            image,
            center_cordinates,
            250,
            (0, 0, 255),  # Red color
            thickness=5
        )
        # Middle Circle (Green, radius 150, thickness 5)
        cv2.circle(
            image,
            center_cordinates,
            150,
            (0, 255, 0),  # Green color
            thickness=5
        )
        # Inner Circle (Blue, radius 50, thickness 5)
        cv2.circle(
            image,
            center_cordinates,
            50,
            (255, 0, 0),  # Blue color
            thickness=-1
        )

        # Diagonal Line (Top-left to Bottom-right)
        cv2.line(
            image,
            (0, 0),
            (image_width, image_height),
            (255, 255, 255),  # White color
            thickness=2
        )

        # horizontal line theought the center
        cv2.line(
            image,
            (0, center_cordinates[1]),
            (image_width, center_cordinates[1]),
            (255, 255, 255),  # White color
            thickness=2
        )

        # adding text to the image
        font_type = cv2.FONT_HERSHEY_SIMPLEX
        text_position = (200, 750)
        cv2.putText(
            image,
            "TARGET",
            text_position,
            font_type,
            1.5,  # Font scale
            (255, 255, 255),  # White color
            thickness=3,
            lineType=cv2.LINE_AA
        )

        #save the image
        save_status = cv2.imwrite("target_image.jpg", image)
        if not save_status:
            raise Exception
        
        # Display the image
        cv2.imshow("Target Image", image)
        cv2.waitKey(0)
        cv2.destroyAllWindows() 

        print("Image created and saved successfully.")

    except Exception as error:
        print(f"An error occurred: {error}")

if __name__ == "__main__":
    main()
```
